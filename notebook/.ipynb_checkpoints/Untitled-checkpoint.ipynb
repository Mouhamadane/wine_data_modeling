```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# File path to the csv files
csv_file_white = "winequality-white.csv"
csv_file_red ="winequality-red.csv"
```

```python
# Read white csv file into dataframe
white_wine = pd.read_csv(csv_file_white, sep=";")

# set white as color
white_wine["color"] = 0

# Print first 5 rows in the dataframe
white_wine.head()
```

```python
# Read red wine csv file into dataframe
red_wine  = pd.read_csv(csv_file_red, sep=";")

# set red as color
red_wine["color"] = 1

# Print first 5 rows in the dataframe
red_wine.head()
```

```python
# inspect the data
red_wine.describe(include="all")
```

```python
white_wine.info()
```

```python
red_wine.info()
```

```python
# merge the red and the white wine dataframes together
wine = red_wine.append(white_wine, ignore_index = True)
```

```python
wine.columns = [x.strip().replace(' ','_') for x in wine.columns]
```

```python
wine.head()
```

```python
wine.info()
```

```python
# set "color" as a categorical variable 
wine['color'] = pd.Categorical(wine['color'])
```

```python
recode = {3: 0, 4: 0, 5: 0, 6: 2, 7: 2, 8: 3, 9: 3}
wine['quality_mark'] = wine['quality'].map(recode)
```

```python
wine.info()
wine.head()
```

```python
target_1=wine.quality_mark
features_1 = wine.drop(['quality','quality_mark'], axis=1)
```

```python
features_1
```

```python
target_1
```

```python
#Splitting into test and train data
from sklearn.model_selection import train_test_split, GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(features_1, target_1, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
l_model = LogisticRegression(max_iter=10000)
l_model.fit(X_train, y_train)
```

```python
l_model.score(X_test, y_test)
```
